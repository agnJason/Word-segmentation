```python
#正向最大匹配
def forwardMM(content, maxlen=8):
    text = content[:]
    answer = []
    
    #当还有剩余字，进行匹配
    while len(text)>0:
        word = text[0:maxlen] #取前maxlen个字匹配
        while len(word)>0:#匹配直到word空了为止
            #当word只剩一个字或者就在字典中时，直接将词放入切分结果中，并text中删除这些词；若不满足，则前向匹配，将word最后一个字去掉继续匹配
            if word in seg_dict or len(word)==1:
                answer.append(word)
                text = text[len(word):]
                break
            else:
                word = word[0:len(word)-1]
    return answer
```

```python
print(forwardMM(content,8))
```

```python
#逆向最大匹配
def backwardMM(content, maxlen=8):
    text = content[:]
    answer = []
    
    #当还有剩余字，进行匹配
    while len(text)>0:
        word = text[len(text)-maxlen:] #取后maxlen个字匹配
        while len(word)>0:#匹配直到word空了为止
            #当word只剩一个字或者就在字典中时，直接将词放入切分结果中，并text中向后删除这些词；若不满足，则后向匹配，将word第一个字去掉继续匹配
            if word in seg_dict or len(word)==1:
                #print(word)
                answer.append(word)
                text = text[:len(text)-len(word)]
                #print(text)
                break
            else:
                word = word[1:]
    answer = [answer[len(answer)-1-n] for n in range(len(answer))]
    return answer
```

```python
print(backwardMM(content))
```
